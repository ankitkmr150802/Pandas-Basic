**Pandas has two types of data structures:**

a) **Series** - It's a one dimensional array with indexes, it stores a single column or row of data in a DataFrame.

b) **Dataframe** - It's a tabular spreadsheet like structure representing rows each of which contains one or
multiple columns.

# loc and iloc

**Loc**

```python
newdf.loc[0,0] = 0.6
newdf.head()
```

```python
newdf.columns = list("abcde")
```

```python
newdf.head()
```

```python
newdf.loc[0,"a"] = 0.8
newdf.head()
```

```python
# incorrect index provided

newdf.loc[0,0] = 1
newdf.head()
```

```python
# drop command

newdf = newdf.drop(0, axis=1)
newdf.head()
```

```python
# to view particular sets of item

newdf.loc[[0,1,3],["a","d"]]
```

```python
# all row 

newdf.loc[:,["a","b"]]
```

```python
# all column

newdf.loc[[0,1,2,3],:]
```

```python
#condition using loc

newdf.loc[newdf["a"] < 0.3]
```

**iloc**

```python
newdf.head()
```

```python
newdf.iloc[0,4]
```

```python
newdf.iloc[[0,1],[0,3]]
```

```python
newdf.drop([1,4],axis=0, inplace = True)
```

```python
newdf.head()
```

```python
newdf.reset_index(drop =True, inplace = True)
```

```python
newdf.head()
```

```python
newdf.shape
```

```python
#adding column 

newdf["f"] = np.random.rand(298)
newdf.head()
```

```python
#made row 0 and row 1 same 

newdf.iloc[1,5] = 0.8
newdf.head()
```

```python
#removes the next same row

newdf.drop_duplicates()
```

```python
#making random column as none

newdf.loc[:,["f"]] = np.nan
newdf.head()
```

```python
newdf["f"].isnull()
```
